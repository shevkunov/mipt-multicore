```python
import numpy as np
import matplotlib.pyplot as plt
import data
```

```python
data = np.array(data.values)
```

```python
plt.figure(figsize=(14, 7))
plt.plot(np.arange(50)[::-1] / 100, data[16], label="Срез в x=0.16")  # на графике из условия перепутаны оси
plt.plot(np.arange(50)[::-1] / 100, data[32], label="Срез в x=0.32")
plt.plot(np.arange(50) / 100, data.T[16], label="Срез в y=0.16") # Только в срезе по x
plt.plot(np.arange(50) / 100, data.T[32], label="Срез в y=0.32")
plt.legend()
plt.show()
```
